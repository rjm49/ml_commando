```python
# Get started with interactive Python
import turtle
from collections import defaultdict

wn = turtle.Screen()
wn.setup(700,700)
wn.bgcolor("black")
wn.colormode(1)

#create pen
class Pen(turtle.Turtle):
  def __init__(self):
    turtle.Turtle.__init__(self)
    self.shape("square")
    self.color("white")
    self.penup()
    self.speed(0)
    
class Player(turtle.Turtle):
  
  def __init__(self):
    turtle.Turtle.__init__(self)
    self.shape("square")
    self.color("yellow")
    self.penup()
    self.speed(0)    
    self.wall_R = -0.75
    self.move_R = -0.04
  
  def go_up(self):
    move_to_x=player.xcor()
    move_to_y=player.ycor() + 24
    
    if (move_to_x,move_to_y) not in walls:
        self.goto(move_to_x, move_to_y)
        return self.move_R
    else:
        return self.wall_R
  
  def go_down(self):
    move_to_x=player.xcor()
    move_to_y=player.ycor() - 24
    
    if (move_to_x,move_to_y) not in walls:
        self.goto(move_to_x, move_to_y)  
        return self.move_R
    else:
        return self.wall_R
        
  def go_right(self):
    move_to_x=player.xcor() + 24
    move_to_y=player.ycor()
    
    if (move_to_x,move_to_y) not in walls:
        self.goto(move_to_x, move_to_y) 
        return self.move_R
    else:
        return self.wall_R
    
  def go_left(self):
    move_to_x=player.xcor() -24
    move_to_y=player.ycor()
    
    if (move_to_x,move_to_y) not in walls:
        self.goto(move_to_x, move_to_y) 
        return self.move_R
    else:
        return self.wall_R

#create Levels list
levels = [""]

#define first level
level_1 = [
"XXXXXXXXXXXXXXXXXXXXXXXXX",
"X                      XX",
"X           XXXXXX  XX XX",
"X  XXXXXXX  XXXXXX  X XXX",
"X       XXXXXXX        XX",
"XXXXXX  XX  XXX        XX",
"XX   X  XX  XXXXXX  XXXXX",
"XX XXX  XX    XXXX  XXXXX",
"X  XXXX       XXXX    XXX",
"X  XXX  XXXXXXXXXXXX   XX",
"X      X  XXXXXXXXXXX   X",
"X  XX    XXX     XXXXX  X",
"X XXXXXXX XX     XXXXX  X",
"X         XXXXX  XXXXX  X",
"XXX  XXXXXXXXXX         X",
"XXX                     X",
"XXXXXXXXXXXXXXXXXXXXXXXXX",
"XXXXXXXXXX     X X     PX",
"X           X           X",
"X   XX XXXXXX  XXXXX  XXX",
"XXXX   XXXX             X",
"X    XX     XXXXXXXXXXX X",
"X XXX   X               X",
"X      GX               X",
"XXXXXXXXXXXXXXXXXXXXXXXXX"
]

#Add maze to mazes list
levels.append(level_1)

#Create Level Setup Function


def setup_maze(level, draw_walls=True, goalstamp=None):
    walls = []
    playable = []
    goal = None
    for y in range(len(level)):
        for x in range(len(level[y])):
          #Get the Character at each x,y coordinate
          #NOTE the order of the y and x in the next line
            character = level[y][x]
    #       Calculate the screen x, y coordinates
            screen_x = -288 + (x * 24)
            screen_y = 288 - (y * 24)

          #Check if it is an X (representing a wall)
            if character == "X":
                if draw_walls:
                    pen.color("white")
                    pen.goto(screen_x, screen_y)
                    pen.stamp()
                walls.append((screen_x,screen_y))
            else:
                playable.append((screen_x, screen_y))

            if character == "P":
                px,py = screen_x, screen_y
                player.goto(screen_x, screen_y)

            if character == "G":
                if goalstamp is not None:
                    pen.clearstamp(goalstamp)
                goal = (screen_x, screen_y)
                pen.color("green")
                pen.goto(screen_x, screen_y)
                goalstamp = pen.stamp()

    return walls, goal, goalstamp, playable, px,py
        
def reset(px,py):
    player.goto(px,py)

def grid_to_screen(gx,gy):
    screen_x = -288 + (gx * 24)
    screen_y = 288 - (gy * 24)
    return screen_x, screen_y

def screen_to_grid(sx,sy):
    gx = (sx+288)//24
    gy = (288-sy)//24
    return gx, gy

def grid_to_onehot(gxy, oh=None):
    gx = gxy[0]
    gy = gxy[1]
    if oh is None:
        oh = numpy.zeros((24,24))
    oh[gx,gy] = 1.0
    return oh
```

```python
from keras import Sequential
from keras.layers import Dense, PReLU
from keras.optimizers import RMSprop, Adam, SGD

model = Sequential()
# model.add(Dense(4, input_shape=(24*24,), activation="linear"))
model.add(Dense(256, input_shape=(576,), activation="relu"))
model.add(Dense(64, activation="relu"))
# model.add(Dense(9, activation="relu"))
model.add(Dense(4))
model.compile(optimizer=Adam(lr=0.001),loss='mse')
```
