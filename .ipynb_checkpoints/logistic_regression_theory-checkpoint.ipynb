```python
%pylab inline

import math
def sigmoid(z):
    return 1.0 / (1.0 + math.exp(-z))

def logit(x):
    return math.log(x/(1-x))
```

```python
number_list = range(-10,10,1)
number_list = sorted(number_list)
sig_list = []
for i in number_list:
    sig = sigmoid(i)
    sig_list.append(sig)
    
logit_list = []
p_list = numpy.arange(0.01,1,0.01)
for i in p_list:
    lg = logit(i)
    logit_list.append(lg)
```

```python
from matplotlib import pyplot as plt
plt.plot(number_list, sig_list)
plt.show()

plt.plot(p_list, logit_list)
plt.show()
```

```python
import numpy as np
def loss_function(y_hat,y):
    L = ( y_hat - y )**2.0
    return L

xs = np.arange(0.0, 1.0, 0.1)
ys = np.arange(0.0, 1.0, 0.1)
zs = [ [loss_function(x,y) for y in ys] for x in xs]
plt.contour(xs,ys,zs)
plt.show()
```

```python
def neg_log_loss_fn(y_hat, y):
    log_loss = y*log(y_hat) + (1.0-y)*(log(1.0-y_hat))
    L = - ( log_loss )
    return L

xs = np.arange(0.001, 0.999, 0.01)
ys = np.arange(0.001, 0.999, 0.01)
zs = [ [neg_log_loss_fn(x,y) for y in ys] for x in xs]
plt.contour(xs,ys,zs)
plt.show()
```

```python
def cost_fn_J(y_hats, ys):
    losses = [ neg_log_loss_fn(y_hat, y) for y_hat,y in zip(y_hats, ys) ]
    sum_loss = sum(losses)
    N = len(losses)
    avg_loss = sum_loss / N
    return avg_loss
```

```python
def linear_regression(xs,ys, n_epochs=1000):
    N = len(xs)
    m = 0.0
    b = 0.0
    learning_rate = 0.01
    
    for i in range(n_epochs):
        cost = 0
        y_hats = []
        for x in xs:
            y_hats.append( b + m*x )
        
        for x,y,y_hat in zip(xs,ys,y_hats):
            cost += (y - y_hat)**2

        dm = 0.0
        db = 0.0    
        for x,y,y_hat in zip(xs,ys,y_hats):  
            dm += -(2/N)*( x * (y - y_hat) )
            db += -(2/N)*( y - y_hat )
            
        m -= learning_rate * dm
        b -= learning_rate * db
    return y_hats, m, b, cost
```

```python
y_hats, m, b, cost = linear_regression(xs,pts)
print(y_hats)
print(m,b)
print(cost)

plt.scatter(xs,pts)
plt.plot(xs,y_hats)
plt.show()
```
