We have a drug that cures some people, but that just makes other people feel worse.
We suspect that expression of gene X/gene Y might be the discriminating factor.
We use LDA to convert from a 2D chart of +ve and -ve examples to a 1D axis that maximises class separation

```python
from sklearn.datasets import make_classification
X,y = make_classification(n_samples=50, n_classes=2, n_features=2, n_informative=2, n_redundant=0, n_repeated=0)

# np.random.seed(666)
fig = plt.figure()
ax = fig.gca()
ax.scatter(X[:,0], X[:,1],c=y)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
lda.fit(X,y)

print(lda.score(X,y))

# fig = plt.figure()
# ax = fig.gca()
print(lda.get_params())
print(lda.means_)

fig = plt.figure()
ax = fig.gca()
ax.scatter(X[:,0], X[:,1],c=y)
ax.scatter(lda.means_[:,0], lda.means_[:,1])
ax.set_xlabel("gene 1")
ax.set_ylabel("gene 2")
plt.show();
```

```python
w = lda.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(min(X[:,0]), max(X[:,0]))
yy = a * xx - (lda.intercept_[0]) / w[1]

fig = plt.figure()
ax = fig.gca()
ax.scatter(lda.means_[:,0], lda.means_[:,1])

ax.scatter(X[:,0], X[:,1],c=y)
ax.plot(xx, yy, 'k--')
plt.ylim( min(X[:,1])-.5, max(X[:,1])+.5)
plt.show();
```
