```python
# split a univariate sequence into samples
def split_sequence(sequence, n_steps):
	X, y = list(), list()
	for i in range(len(sequence)):
		# find the end of this pattern
		end_ix = i + n_steps
		# check if we are beyond the sequence
		if end_ix > len(sequence)-1:
			break
		# gather input and output parts of the pattern
		seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
		X.append(seq_x)
		y.append(seq_y)
	return array(X), array(y)


# numpy.set_printoptions(precision=2, suppress=True)
# tX,ty = split_sequence(univariate, n_steps)
# print(len(tX))
# for x,y in zip(tX,ty):
#     print("{} {:.2f}".format(x,y))

ts_fwd = 0
ts_bwd = 1
n_steps = ts_fwd+ts_bwd

df = pandas.DataFrame(columns=["t"], data=univariate)
df["t"]
for i in range(ts_fwd):
    df.insert(0, "t+{}".format((i+1)), df["t"].shift(-(i+1)))
for i in range(ts_bwd):
    df["t-{}".format(i+1)]=df["t"].shift(i+1)

print(df.columns)
# df.dropna(inplace=True)
df = df[df.columns[::-1]]
print(df)
df = df.values

tX = xs[0:-split_at]
vX = xs[-split_at:]
# tX = df[0:-split_at, 0]
# vX = df[-split_at:, 0]
ty = df[0:-split_at, -1]
vy = df[-split_at:, -1]

for x,y in zip(tX, ty):
    print(x,y)

n_features = 1
```
